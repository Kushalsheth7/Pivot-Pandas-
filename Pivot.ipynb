```python
import pandas as pd


data = {
    'Product': ['A', 'B', 'C', 'A', 'B', 'C'],
    'Region': ['North', 'North', 'North', 'South', 'South', 'South'],
    'Sales': [100, 200, 150, 300, 400, 350]
}

df = pd.DataFrame(data)
print(df)


pivot_df = df.pivot(index='Region', columns='Product', values='Sales')

print(pivot_df)
```

```python

data = {
    'Product': ['A', 'B', 'C', 'A', 'B', 'C', 'A', 'B', 'C'],
    'Region': ['North', 'North', 'North', 'South', 'South', 'South', 'East', 'East', 'East'],
    'Sales': [100, 200, 150, 300, 400, 350, 500, 600, 550]
}

df = pd.DataFrame(data)
print(df)


pivot_table_df = df.pivot_table(values='Sales', index='Region', columns='Product', aggfunc='sum', fill_value=0)

print(pivot_table_df)
```

```python

data = {
    'Product': ['A', 'B', 'C', 'A', 'B', 'C', 'A', 'B', 'C'],
    'Region': ['North', 'North', 'North', 'South', 'South', 'South', 'East', 'East', 'East'],
    'Sales': [100, 200, 150, 300, 400, 350, 500, 600, 550],
    'Profit': [10, 20, 15, 30, 40, 35, 50, 60, 55]
}

df = pd.DataFrame(data)
print(df)

pivot_table_df = df.pivot_table(values=['Sales', 'Profit'], index='Region', columns='Product', aggfunc={'Sales': 'sum', 'Profit': 'mean'}, fill_value=0)

print(pivot_table_df)
```
